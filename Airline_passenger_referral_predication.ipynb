````python
columns_to_impute = ['seat_comfort', 'cabin_service', 'food_bev',
                     'entertainment', 'ground_service', 'value_for_money']
# Impute missing values with median
for column in columns_to_impute:
    median_value = airline_df[column].median()  # Calculate median
    airline_df[column].fillna(median_value, inplace=True)  # Replace missing values with median

# Convert columns to int64
for column in columns_to_impute:
    airline_df[column] = airline_df[column].astype(int)  # Convert to int64 data type
````

data follows a symmetric distribution and doesn't have strong outliers,we have the flexibility to choose between mean and median imputation. we are imputing median values and changing datatype float to interger is a common preprocessing step for preparing dataset for machine learning.

````python
# List of columns with missing values of type 'object'
columns_to_impute = ['airline','author','review_date','customer_review','aircraft', 'cabin','traveller_type', 'route', 'date_flown','recommended']

# Impute missing values with mode
for column in columns_to_impute:
    mode_value = airline_df[column].mode()[0]  # Calculate mode
    airline_df[column].fillna(mode_value, inplace=True)  # Replace missing values with mode
````

## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

Hypothesis Question 1:
Are passengers who travel in Business Class more likely to give higher ratings for seat comfort compared to passengers in Economy Class?

Hypothesis Question 2:
Does the overall rating of flights vary significantly between different aircraft types?

Hypothesis Question 3:
Hypothesis Question 3:
Is there a significant difference in the overall ratings provided by passengers who recommend the airline and those who do not recommend the airline?

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Null Hypothesis (H0):μ_business = μ_economy
There is no significant difference in the average seat comfort ratings between passengers who travel in Business Class and passengers who travel in Economy Class.

Alternative Hypothesis (H1):μ_business > μ_economy
Passengers who travel in Business Class are more likely to give higher average seat comfort ratings compared to passengers who travel in Economy Class.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
# Define the ratings for Business Class and Economy Class
ratings_business = airline_df[airline_df['cabin'] == 'Business Class']['seat_comfort']
ratings_economy = airline_df[airline_df['cabin'] == 'Economy Class']['seat_comfort']

# Perform a t-test for independent samples
t_statistic, p_value = stats.ttest_ind(ratings_business, ratings_economy, equal_var=False)

# Print the results
print("t-statistic:", t_statistic)
print("p-value:", p_value)
````


The obtained p-value is 0.0, which is significantly lower than the common significance level of 0.05. In hypothesis testing, a p-value below the significance level suggests that reject the null hypothesis.

The t-statistic value of 53.896 further supports this conclusion by indicating a large difference in means between the two groups.

Therefore, we would reject the null hypothesis and accept the alternative hypothesis, concluding that there is indeed a significant difference in seat comfort ratings between Business Class and Economy Class passengers. This outcome could have implications for airline service improvements, targeted marketing strategies, or other business decisions to enhance passenger experience.






##### Which statistical test have you done to obtain P-Value?

I performed an independent samples t-test to obtain the p-value. This statistical test is used to compare the means of two independent groups to determine if there is a significant difference between them. In this specific case, I used the t-test to compare the seat comfort ratings of Business Class and Economy Class passengers. The obtained p-value of 0.0 indicates that there is a significant difference in seat comfort ratings between the two groups.

##### Why did you choose the specific statistical test?

I chose the independent samples t-test because it is appropriate when comparing the means of two independent groups. In this scenario, we are comparing the seat comfort ratings of two different passenger classes (Business Class and Economy Class), and we want to determine if there is a significant difference between their ratings. The t-test is commonly used for such comparisons and is suitable when the data follows a roughly normal distribution and the groups are independent.

### Hypothetical Statement - 2

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Null Hypothesis (H0):
H0: μ1 = μ2 (There is no significant difference in the overall ratings between different aircraft types.)

Alternate Hypothesis (Ha):
Ha: μ1 ≠ μ2 (There is a significant difference in the overall ratings between different aircraft types.)

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
# Data for each aircraft type
aircraft1_ratings = airline_df[airline_df['aircraft'] == 'A320']['overall']
aircraft2_ratings = airline_df[airline_df['aircraft'] == 'A380']['overall']
# Add more aircraft types as needed

# Perform ANOVA test
f_statistic, p_value = stats.f_oneway(aircraft1_ratings, aircraft2_ratings)
alpha = 0.05  # Set the significance level

print("F-Statistic:", f_statistic)
print("P-Value:", p_value)

if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference in overall ratings between aircraft types.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in overall ratings between aircraft types.")
````

Reject the null hypothesis. The p-value (3.61e-60) is significantly smaller than the common significance level (alpha) typically chosen for hypothesis tests (e.g., 0.05). This indicates that the probability of obtaining such extreme results under the assumption that there is no difference in overall ratings between aircraft types is extremely low. Therefore, you have strong evidence to reject the null hypothesis and conclude that there is a significant difference in overall ratings between different aircraft types.

##### Which statistical test have you done to obtain P-Value?

The statistical test that was performed to obtain the p-value is the Analysis of Variance (ANOVA) test. The ANOVA test is used to compare means across multiple groups to determine if there are statistically significant differences among the group means. In this case, it was used to compare the overall ratings of different aircraft types to see if there is a significant difference in ratings among the groups. The obtained p-value from the ANOVA test helps determine whether to reject or fail to reject the null hypothesis based on a pre-defined significance level (alpha)

##### Why did you choose the specific statistical test?

The specific statistical test, Analysis of Variance (ANOVA), was chosen because it is suitable for comparing the means of more than two groups, which is the case when we want to analyze the differences in overall ratings among different aircraft types. ANOVA allows us to determine if there are statistically significant differences among the group means while taking into account the variance within each group.

In this context, we are comparing the overall ratings of different aircraft types, and ANOVA can help us understand whether any observed differences in means are likely due to true differences or are simply due to random variability. Since we are dealing with multiple groups (aircraft types), ANOVA is a powerful tool for such comparisons.

This means that none of the data points in the given DataFrame were found to be significantly distant from the mean values in these columns based on the chosen threshold.

The Z-score method is used to detect outliers in the provided numeric columns of the DataFrame. This method calculates the Z-score for each data point, indicating how many standard deviations it is away from the mean of that column. #### 2. Feature Selection

````python
from sklearn.feature_selection import SelectKBest, f_classif
# Separate the feature matrix 'X' and the target variable 'y'
X = airline_df.drop(columns=['recommended'])
y = airline_df['recommended']

# Number of top features to select
k = 8

# Perform feature selection using ANOVA
selector = SelectKBest(score_func=f_classif, k=k)
X_selected = selector.fit_transform(X, y)

# Get the indices of the selected features
selected_feature_indices = selector.get_support(indices=True)

# Get the names of the selected features
selected_feature_names = X.columns[selected_feature_indices]
selected_feature_scores = selector.scores_[selected_feature_indices]

# Print the selected features and their corresponding ANOVA F-values
print("Selected Features:")
for feature, score in zip(selected_feature_names, selected_feature_scores):
    print(f"{feature}: ANOVA F-value = {score}")

````

The purpose of this code is to select relevant features that are most correlated with the target variable ('recommended') based on their ANOVA F-values.
ANOVA (Analysis of Variance) measures the variance between multiple groups to determine whether there are significant differences among those groups.
In this case, the code is attempting to identify which features have the highest influence on the 'recommended' column, which could be used to predict whether a review is recommended or not.

By using ANOVA F-values, the code identifies which features have a statistically significant impact on the target variable.
The selected features and their ANOVA F-values are printed to help you decide which features are important for your modeling task.

##### What all feature selection methods have you used  and why?

SelectKBest Algorithm:

The SelectKBest algorithm is a technique for feature selection that helps to reduce the number of features while retaining the most informative ones.
It selects the top 'k' features based on a scoring function that measures the importance of each feature with respect to the target variable.
ANOVA F-Value Scoring Function:

In our case, we used the ANOVA F-value scoring function as the metric to evaluate the importance of each feature.
ANOVA (Analysis of Variance) is a statistical method used to analyze differences among group means in a sample.
The ANOVA F-value quantifies the ratio of between-group variance to within-group variance. A higher F-value suggests a stronger association between the feature and the target variable.
